#### This Notebook implements the CF using the PCP procedures as input and Spesialist procedures as output, based on Autoencoder implemented by Keras

## Load the data

```python
import numpy as np
import pandas as pd

data_test = pd.read_csv('data_processing/data_test.csv')
data_train = pd.read_csv('data_processing/data_train.csv')



# the data is in the form of feature matrix, so can directly be fed into the NN
PC_pr_tr = data_train[[data_train.columns[0]]+ list(data_train.columns[314: 375])].sort_values('PC_enc')
PC_pr_ts = data_test[[data_test.columns[0]]+ list(data_test.columns[314: 375])].sort_values('PC_enc')
print(PC_pr_tr.head())

SP_pr_tr = data_train[[data_train.columns[0]]+ list(data_train.columns[407:])].sort_values('PC_enc')
SP_pr_ts = data_test[[data_test.columns[0]]+ list(data_test.columns[407:])].sort_values('PC_enc')
print(SP_pr_tr.head())




X_train =  np.logical_or(PC_pr_tr.to_numpy()[:,1:], SP_pr_tr.to_numpy()[:,1:])
X_test =  PC_pr_ts.to_numpy()[:,1:] #logical_or(PC_pr_ts.to_numpy()[:,1:], SP_pr_ts.to_numpy()[:,1:])

d = X_train.shape[1]
#Y_train = SP_pr_tr.to_numpy()[:,1:]
#Y_test = SP_pr_ts.to_numpy()[:,1:]
```

### Process data. 

## Model: AE

```python
from keras.layers import Input, Dense, Dropout
from keras.models import Model

p = 0.3

input_data = Input(shape=(d,))
encoded = Dense(60, activation='relu')(input_data)
encoded = Dropout(rate=p)(encoded)
encoded = Dense(40, activation='relu')(encoded)
encoded = Dropout(rate=p)(encoded)
decoded = Dense(60, activation='relu')(encoded)
encoded = Dropout(rate=p)(encoded)
decoded = Dense(d, activation='sigmoid')(decoded)

#encoded = Dense(40, activation='relu')(input_img)
#decoded = Dense(d, activation='sigmoid')(encoded)

#encoding_dim = 40

# this model maps an input to its reconstruction
autoencoder = Model(input_data, decoded)
```

#### Now let's train our autoencoder

```python
autoencoder.compile(optimizer='sgd', loss = 'mean_squared_error') #loss='mean_squared_error')
```

#### Now let's train our autoencoder for 50 epochs:

```python

autoencoder.fit(X_train, X_train,
                epochs=500,
                batch_size=256,
                shuffle=True,
                validation_data=(X_test, X_test))#,class_weight=weights) 
```

#### After 50 epochs, the autoencoder seems to reach a stable train/test loss value of about 0.11. We can try to visualize the reconstructed inputs and the encoded representations. We will use Matplotlib.

```python
# encode and decode some digits
# note that we take them from the *test* set

#encoded_imgs = encoder.predict(x_test)
#decoded_imgs = decoder.predict(encoded_imgs)
# output of the collaborative filtering
pred = autoencoder.predict(X_test)

# predictions for the test dataset: subtract pred from the 
pred[np.where(PC_pr_ts.to_numpy()[:,1:]==1)]=0 

# save the pred results for train data:
pred_train = autoencoder.predict(X_train)

# predictions for the test dataset: subtract pred from the 
pred_train[np.where(PC_pr_tr.to_numpy()[:,1:]==1)]=0 
```
